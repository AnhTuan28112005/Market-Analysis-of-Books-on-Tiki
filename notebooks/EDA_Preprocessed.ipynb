```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load dữ liệu thô
df = pd.read_csv('../data/raw/tiki_books_raw.csv', encoding='utf-8')
print(f"Shape: {df.shape}")
print(f"\nCác cột: {df.columns.tolist()}")
print(f"\nKiểu dữ liệu:\n{df.dtypes}")
```

```python
print("\n--- 2. Kiểm tra giá trị thiếu (Missing values) ---")
missing_count = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Count': missing_count, 'Percentage (%)': missing_percentage})
print(missing_df[missing_df['Count'] > 0])

print("\n--- 3. Kiểm tra trùng lặp (Duplicates) ---")
print(f"Số dòng trùng lặp hoàn toàn: {df.duplicated().sum()}")
print(f"Số dòng trùng lặp theo product_id: {df.duplicated(subset=['product_id']).sum()}")
```

```python
print("--- 4. Thống kê mô tả cột số ---")
display(df.describe())

# Kiểm tra xem có sách nào giá bằng 0 hoặc lượt bán âm không (vô lý)
print("\n--- 5. Kiểm tra tính hợp lệ ---")
print(f"Sách có giá <= 0: {len(df[df['price'] <= 0])}")
print(f"Sách có list_price < price: {len(df[df['list_price'] < df['price']])}")
```

```python
# 1. Lọc ra tất cả các dòng có product_id bị trùng lặp
# keep=False giúp hiển thị cả "bản gốc" và các "bản sao" của nó
duplicate_df = df[df.duplicated(subset=['product_id'], keep=False)]

# 2. Sắp xếp theo product_id để các dòng trùng nhau hiện ngay sát nhau cho dễ so sánh
duplicate_df_sorted = duplicate_df.sort_values(by='product_id')

# 3. In ra kết quả (ví dụ xem 20 dòng đầu liên quan đến trùng lặp)
print(f"Tìm thấy {len(duplicate_df_sorted)} dòng có ID bị trùng lặp.")
display(duplicate_df_sorted.head(20))
```

```python
# 1. Lọc ra danh sách các cột có kiểu dữ liệu là số (int, float)
numeric_cols = df.select_dtypes(include=['number']).columns

# 2. Kiểm tra xem từng cột đó có phải tất cả giá trị đều bằng 0 không
# (df[col] == 0).all() sẽ trả về True nếu mọi hàng trong cột đều là số 0
zero_cols = [col for col in numeric_cols if (df[col] == 0).all()]

# 3. In kết quả
print("--- KIỂM TRA CỘT TOÀN GIÁ TRỊ 0 ---")
if len(zero_cols) > 0:
    print(f"Tìm thấy {len(zero_cols)} cột có giá trị toàn bằng 0:")
    for col in zero_cols:
        print(f" - {col}")
else:
    print("Không có cột số nào có giá trị toàn bằng 0. ✅")

# Mở rộng: Kiểm tra xem có cột nào bị toàn giá trị thiếu (NaN) không
nan_cols = [col for col in df.columns if df[col].isnull().all()]
if nan_cols:
    print(f"\nCảnh báo: Các cột toàn giá trị NaN: {nan_cols}")
```

```python
# Sắp xếp theo doanh số giảm dần (để ưu tiên giữ lại dòng có doanh số được ghi nhận cao nhất - nếu có xê dịch lúc cào)
df = df.sort_values(by='quantity_sold', ascending=False)

# Xóa các dòng trùng product_id, chỉ giữ lại dòng xuất hiện đầu tiên
df = df.drop_duplicates(subset=['product_id'], keep='first').reset_index(drop=True)

print(f"Số dòng sau khi xóa trùng lặp: {len(df)}")
```

```python
# 1. Điền giá trị cho loại bìa và người bán bị thiếu
df['cover_type'] = df['cover_type'].fillna('Không rõ')
df['seller_name'] = df['seller_name'].fillna('Chưa rõ người bán')

# 2. Loại bỏ các giá trị nhiễu ở cột số như None, NaN
# Ở bước trước `image_count` có thể bị rỗng, điền 0
df['image_count'] = df['image_count'].fillna(0).astype(int)

# 3. Kiểm tra lại lần chót
print("Số dòng missing còn lại (chỉ tính cột quan trọng):")
print(df[['cover_type', 'seller_name', 'quantity_sold', 'price', 'image_count']].isnull().sum())
```

```python
print("--- Trước khi lọc Outlier ---", len(df))

# 1. Xóa các sách có giá bằng 0 hoặc bé hơn 0 (Lỗi hệ thống)
# 2. Xóa sách giá quá 5.000.000đ (Thường không phải sách mà là gộp combo tạp nham)
df = df[(df['price'] > 0) & (df['price'] <= 5_000_000)].reset_index(drop=True)

print("--- Sau khi lọc Outlier ---", len(df))
print(f"Giá phân bổ từ: {df['price'].min():,}đ đến {df['price'].max():,}đ")
```

```python
# Tạo cột phân nhóm số lượng hình ảnh (Phục vụ Mục Tiêu 4.1 của TV4)
def classify_images(img_count):
    if img_count <= 1:
        return '1 ảnh (Thô sơ)'
    elif img_count <= 3:
        return '2-3 ảnh'
    elif img_count <= 5:
        return '4-5 ảnh'
    else:
        return '>5 ảnh (Chi tiết)'

df['image_group'] = df['image_count'].apply(classify_images)

# Kiểm tra thử cột mới
print("Phân bố sách theo nhóm hình ảnh trên Tiki:")
print(df['image_group'].value_counts())
```
